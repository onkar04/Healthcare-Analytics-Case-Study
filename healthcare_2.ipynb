```python
import pandas as pd

#train=pd.read_csv("readmission_train.csv")
```

```python
df1=pd.read_csv("concatdf2.csv", low_memory=False)
```

```python
df1['num_times_admitted'] = df1.number_outpatient + df1.number_inpatient + df1.number_emergency
```

```python
df1['num_procedures1'] = df1.num_procedures + df1.num_lab_procedures
```

```python
df1['lab_percentage'] = df1.num_lab_procedures*100/df1.num_procedures1
df1['out_percent'] = df1.number_outpatient*100/df1.num_times_admitted
df1['in_percent'] = df1.number_inpatient*100/df1.num_times_admitted
df1['emergency_percent'] = df1.number_emergency*100/df1.num_times_admitted
```

```python
df1.head()
```

```python
df2 = df1[['diag_1','diag_2','diag_3']]
```

```python
df2.head()
```

```python
df2['diag_1_cat'] = df2['diag_1'][df2['diag_1'].str.contains('[A-Za-z]')]
df2['diag_2_cat'] = df2['diag_2'][df2['diag_2'].str.contains('[A-Za-z]')]
df2['diag_3_cat'] = df2['diag_3'][df2['diag_3'].str.contains('[A-Za-z]')]
```

```python
df2['diag_1_cat'].fillna('Missing', inplace = True)
df2['diag_2_cat'].fillna('Missing', inplace = True)
df2['diag_3_cat'].fillna('Missing', inplace = True)
```

```python
df2['diag_1'][df2['diag_1'].str.contains('[A-Za-z]')] = 0
df2['diag_2'][df2['diag_2'].str.contains('[A-Za-z]')] = 0
df2['diag_3'][df2['diag_3'].str.contains('[A-Za-z]')] = 0
```

```python
df2.head()
```

```python
df3 = pd.concat([df1,df2], axis=1)
```

```python
df3.head()
```

```python
df3.to_csv('featured_new.csv')
```
